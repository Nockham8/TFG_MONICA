## 2. Ejecución face_recognition


Creamos un dataframe de la lista de txt de todas las imágenes que vamos a utilizar.

```python
dt = pd.read_csv('/content/drive/My Drive/TFG/BLOQUE_1/CelebA/Img/nombres_de_imagenes.txt', sep=' ', header=None)
dt.head()
```

```python
dt
```

Aplicamos face_recogition

```python
show = False
for nom in dt[0]:        ## totes
#for nom in dt[0][:10]:  ## images de la 0 a la 9
  imageFR = face_recognition.load_image_file('/content/drive/My Drive/TFG/BLOQUE_1/CelebA/Img/main_img/'+nom)
  face_locations = face_recognition.face_locations(imageFR, number_of_times_to_upsample=0, model="cnn")
  if show:
    print("I found {} face(s) in this photograph.".format(len(face_locations)))
  image = cv2.imread('/content/drive/My Drive/TFG/BLOQUE_1/CelebA/Img/main_img/'+nom)
  pil_image = Image.fromarray(image)
  draw = ImageDraw.Draw(pil_image)
  for top, right, bottom, left in face_locations:
    crop_image = image[top:bottom, left:right]
    if show:
      cv2_imshow(image)
      cv2_imshow(crop_image)
    cv2.imwrite('/content/drive/My Drive/TFG/BLOQUE_1/CelebA/Img/cut_faces/'+nom, crop_image)
    break
```
